```python
import matplotlib.pyplot as plt
import csv

재수생 = []
현역 = []
year = []

with open('data.csv', 'r', encoding='cp949') as reader_csv:
    reader = csv.reader(reader_csv, delimiter=',')
    next(reader)
    for row in reader:
        year.append(row[0])
        재수생.append(int(row[10]))
        현역.append(int(row[9]))


#plt.plot(year,data,color='green',marker='o',linestyle='solid')
#plt.plot([1,3,2,5])
plt.rc('font',family='Malgun Gothic')
plt.figure(dpi=300,figsize=(20,5))
plt.plot(year,재수생)
plt.plot(year,현역)

plt.show()
```

```python
names = []
datas = []
titles = []

def time(val):
    return val[:2]

p = re.compile('\(.*?\)')

with open('subway.csv', 'r', encoding='cp949') as reader_csv:
    reader = csv.reader(reader_csv, delimiter=',')
    titles = list(next(reader)[3:-1][0::2])
    subway_name = input("지하철 역 이름 입력: ")
    
    for row in reader:
        if (re.sub(p, "", row[2]) == subway_name):
            datas = list(map(int,row[3:-1][0::2])) 
            break
    

if (len(datas) == 0):
    print("정확한 역 이름을 입력해주세요")
else:
    plt.rc('font',family='Malgun Gothic')
    plt.figure(dpi=300,figsize=(20,5))
    plt.plot(titles,datas)

    plt.show()
```
