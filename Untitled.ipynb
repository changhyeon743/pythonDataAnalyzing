```python
names = []
datas = []
titles = []

def time(val):
    return val[:2]

p = re.compile('\(.*?\)')

with open('subway.csv', 'r', encoding='cp949') as reader_csv:
    reader = csv.reader(reader_csv, delimiter=',')
    titles = list(next(reader)[3:-1][0::2])
    subway_name = input("   : ")
    
    for row in reader:
        if (re.sub(p, "", row[2]) == subway_name):
            datas = list(map(int,row[3:-1][0::2])) 
            break
    

if (len(datas) == 0):
    print("   ")
else:
    plt.rc('font',family='Malgun Gothic')
    plt.figure(dpi=300,figsize=(20,5))
    plt.plot(titles,datas)

    plt.show()
```
